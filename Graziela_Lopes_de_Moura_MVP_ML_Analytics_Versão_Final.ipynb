### 1.1 O Desafio de Prever os Gastos com a Saúde

O setor de saúde suplementar trabalha diariamente com a imprevisibilidade financeira decorrente da sinistralidade, compreendida entre as despesas assistenciais (custos com consultas, exames, internações e cirurgias) e a receita total arrecadada com as mensalidades. A variação desse indicador é fortemente influenciada por fatores demográficos — tais como idade e região de residência — e por hábitos de vida e características clínicas, como o tabagismo e o Índice de Massa Corporal (IMC). Existem diversos estudos que comprovam que estas variáveis, hábitos de vida, impactam na frequência e a gravidade dos atendimentos médicos demandados.

A falta de ferramentas de inteligência para prever despesas impede que as operadoras saibam quanto sua carteira de clientes vai custar no futuro. Essa incerteza constante atrapalha a administração e o controle dos da empresa.

Com o propósito de reduzir essa lacuna na operação, este projeto visa o desenvolvimento de um modelo de Machine Learning voltado para a **regressão e previsão** do custo médico anual individual de novos beneficiários, atuando logo no ingresso do cliente na carteira de seguros. A automação dessa estimativa fornecerá subsídios fundamentais para três pilares de decisão estratégica de negócios:

* **Precificação de Contratos:** Viabiliza a definição de valores de mensalidades de forma atuarialmente justa, equilibrando os gastos esperados com os atendimentos médicos (a sinistralidade) com um preço que seja justo para o cliente pagar e, ao mesmo tempo, competitivo para a empresa no mercado.
* **Provisão Financeira:** Ajuda o setor financeiro a prever o valor exato que deve ser mantido em caixa como garantia. Dessa forma, a empresa nunca é pega de surpresa e tem sempre o dinheiro necessário para cobrir os futuros gastos médicos dos seus clientes.
* **Programas de Prevenção e Promoção da Saúde:** Identifica precocemente perfis com alta probabilidade de alto custo, direcionando-os de forma preventiva a programas de bem-estar e gestão de saúde crônica, como grupos de combate ao tabagismo ou acompanhamento nutricional direcionado.

A **relevância desse projeto** se deve ao fato de que, no cenário atual, gerenciar o custo dos serviços médicos se tornou o maior desafio econômico para as empresas de saúde privada.

Isso sugere que o IMC de 30 possa funcionar como um gatilho, mas não age sozinho. É porvavel que outra variável que, combinada à obesidade, faça o custo decolar.

O código abaixo dividiu a tela em duas partes idênticas: Mulheres e Homens. Dentro de cada lado, os pacientes foram pintados em três cores:

Verde (Peso Normal): Pacientes saudáveis (IMC < 25).

Amarelo (Sobrepeso): Pacientes na zona de alerta (IMC entre 25.1 e 29.9).

Vermelho (Obesidade): Pacientes na zona de risco clínico (IMC >= 30).

Para averiguar se o genero não é realmente o fator relevante para o custo.

```python
# --- Gráfico 5: Visão Global - Idade vs Custo por Gênero e Categorias ABRAN ---

def categorizar_imc_abran(imc):
    if imc < 25.0:
        return "1. Normal/Baixo Peso (<25)"
    elif imc < 30.0:
        return "2. Sobrepeso (25-29.9)"
    else:
        return "3. Obesidade (>=30)"

# Criando a coluna de categorias da ABRAN na base principal
df["Categoria_IMC"] = df["bmi"].apply(categorizar_imc_abran)

# Gerando o gráfico cruzado completo
g = sns.lmplot(
    data=df.sort_values("Categoria_IMC"), # Ordena para a legenda ficar certinha
    x="age",
    y=TARGET,
    hue="Categoria_IMC",
    col="sex",                      # Telas separadas para Homens e Mulheres
    palette={"1. Normal/Baixo Peso (<25)": "#2ecc71",  # Verde (Saudável)
             "2. Sobrepeso (25-29.9)": "#f1c40f",      # Amarelo (Alerta)
             "3. Obesidade (>=30)": "#e74c3c"},        # Vermelho (Risco Clínico)
    height=5,
    aspect=1.2,
    scatter_kws={"alpha": 0.6},
    ci=None # Remove a sombra da linha para o gráfico ficar mais limpo
)

# Ajustando títulos e eixos
g.set_xlabels("Idade")
g.set_ylabels("Custos Médicos ($)")
g.fig.subplots_adjust(top=0.8)
g.fig.suptitle("Análise Macroglobal: Impacto das Faixas de IMC (ABRAN) e Idade por Gênero", fontsize=14)

plt.show()
```

O mapeamento baseado nas diretrizes da ABRAN revela três conclusões sobre o comportamento dos custos:

# 5. Preparação dos dados e divisão treino/teste

A preparação da base consistiu na separação das variáveis e na segmentação dos dados para o aprendizado do modelo:

Features e Target: Isolamos a coluna charges como a variável alvo (target) e mantivemos as demais colunas (age, sex, bmi, children, smoker, region) como preditoras (features). Nenhuma coluna foi descartada, pois a análise exploratória comprovou a relevância de todas elas.

Proporção da Divisão: Os dados foram divididos na proporção de 80% para treino e 20% para teste. Essa escolha garante um volume robusto de dados para o aprendizado dos algoritmos e uma amostra estatisticamente confiável para validação com dados inéditos.

Estratégia de Divisão: Como o problema não é temporal (não envolve séries cronológicas), os dados foram embaralhados aleatoriamente antes da divisão. Fixado o parâmetro de reprodutibilidade (random_state) para garantir que os mesmos grupos sejam gerados em futuras execuções.

```python
# === Configuração do problema ===

TARGET = "charges"                # o alvo é o custo médio por enquabto
PROBLEM_TYPE = "regressao"        # "regressao"
ID_COLUMNS = []                    # não tem colunas de ID
DATE_COLUMN = None                 # Não é uma série temporal (dados transversais)
DROP_COLUMNS = []                  # Mantive todas as features pois são relevantes

if PROBLEM_TYPE != "clusterizacao":
    assert TARGET is not None and TARGET in df.columns, "Defina um TARGET válido."

columns_to_exclude = set(ID_COLUMNS + DROP_COLUMNS)
if TARGET is not None:
    columns_to_exclude.add(TARGET)

features = [c for c in df.columns if c not in columns_to_exclude]
print("Tipo de problema:", PROBLEM_TYPE)
print("Target:", TARGET)
print("Número de features:", len(features))
print("Features:", features)
```

```python
# === Divisão dos dados ===

if PROBLEM_TYPE in ["classificacao", "regressao"]:
    X = df[features].copy()
    y = df[TARGET].copy()

    stratify_arg = y if PROBLEM_TYPE == "classificacao" else None
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=SEED, stratify=stratify_arg
    )

    print("Treino:", X_train.shape, "| Teste:", X_test.shape)

elif PROBLEM_TYPE == "clusterizacao":
    X = df[features].copy()
    X_train, X_test = train_test_split(X, test_size=0.2, random_state=SEED)
    print("Treino:", X_train.shape, "| Teste:", X_test.shape)

elif PROBLEM_TYPE == "serie_temporal":
    assert DATE_COLUMN is not None and DATE_COLUMN in df.columns, "Defina DATE_COLUMN para séries temporais."
    df_sorted = df.sort_values(DATE_COLUMN).copy()
    cutoff = int(len(df_sorted) * 0.8)
    train_df = df_sorted.iloc[:cutoff]
    test_df = df_sorted.iloc[cutoff:]

    X_train = train_df[features].copy()
    y_train = train_df[TARGET].copy()
    X_test = test_df[features].copy()
    y_test = test_df[TARGET].copy()

    print("Treino:", X_train.shape, "| Teste:", X_test.shape)

else:
    raise ValueError("PROBLEM_TYPE inválido.")
```


5.1 Justificativa da divisão
A divisão por Holdout se justifica porque os dados não dependem do tempo (não é uma linha do tempo ou histórico de meses/anos), então o método tradicional de separar uma parte para treino e outra para teste funciona perfeitamente. É a abordagem simples e eficiente para esse tipo de base.

Separar 80% dos dados para treino e 20% para teste é o equilíbrio ideal para o tamanho do dataset (cerca de 1.300 pacientes). Isso dá ao computador exemplos suficientes para ele aprender as regras do plano de saúde (cerca de 1.070 pessoas) e reserva uma quantidade justa (cerca de 267 pessoas) para testarmos se ele realmente aprendeu as respostas.

O objetido deste trabalho é prever um valor que muda continuamente (o custo em dinheiro) por isso não houve a necessidade de estratificação. Além disso, vimos na análise anterior que os grupos (homens/mulheres, regiões) já são bem equilibrados na base.

Proteção Contra Vazamento de Dados: Ao separar os 20% de teste antes de fazer qualquer transformação nos dados, garantimos que o modelo não tenha nenhuma pista ou "espie" informações do grupo de teste durante o seu aprendizado. Isso evita que o modelo pareça perfeito nos testes, mas falhe na vida real.


### 6 Decisões de pré-processamento

* **Escolha da Imputação (Média/Mediana/Moda):**
  Como nossa análise exploratória mostrou que o dataset está 100% preenchido, a imputação não afetará os dados atuais. Contudo, deixamos configurada a **mediana** para os atributos numéricos e a **moda (mais frequente)** para os categóricos como uma trava de segurança para o futuro. Escolhemos a mediana em vez da média porque ela é muito mais robusta a *outliers* (valores discrepantes), garantindo que, se um paciente com idade ou IMC bizarro aparecer com dado nulo no futuro, o modelo não seja distorcido.

* **Padronização ou Normalização dos Dados:**
  Utilizamos a **padronização (`StandardScaler`)** para todas as variáveis numéricas (`age`, `bmi`, `children`). Essa escolha é crucial porque as nossas variáveis estão em ordens de grandeza completamente diferentes (idades até 64, filhos até 5, IMC até 53). Ao transformar esses dados para que todos tenham média 0 e desvio padrão 1, garantimos que algoritmos baseados em distância ou penalização não fiquem enviesados dando peso maior para a variável com os maiores números absolutos.

* **Variáveis Removidas e o Porquê:**
  Nenhuma variável original do dataset foi removida por irrelevância, pois todas as 6 colunas (`age`, `sex`, `bmi`, `children`, `smoker` and `region`) possuem forte fundamentação teórica e epidemiológica na literatura médica para o cálculo de risco de sinistro. A única remoção realizada foi a da **1 linha duplicada**, o que é uma limpeza de integridade técnica e não uma exclusão de atributos.

* **Criação de Novas Variáveis (Engenharia de Features):**
  Nenhuma coluna precisou ser excluída ou criada. A análise inicial mostrou que todas as informações originais da base são relevantes e suficientes para prever o custo do paciente de forma clara.


```python
# Identificação automática simples de colunas numéricas e categóricas.

num_cols = X_train.select_dtypes(include=np.number).columns.tolist()
cat_cols = X_train.select_dtypes(exclude=np.number).columns.tolist()

# Remover da coluna de data (caso existisse)
if DATE_COLUMN in cat_cols:
    cat_cols.remove(DATE_COLUMN)
if DATE_COLUMN in num_cols:
    num_cols.remove(DATE_COLUMN)

#Pipeline para variáveis numéricas (Idade, IMC, Filhos)
numeric_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])
#Pipeline para variáveis categóricas (Genero, Fumantes, Região)
categorical_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])
#Combinando os dois tratamentos em um único objeto de pré-processamento
preprocess = ColumnTransformer(transformers=[
    ("num", numeric_pipe, num_cols),
    ("cat", categorical_pipe, cat_cols)
], remainder="drop")

print("Colunas numéricas:", num_cols)
print("Colunas categóricas:", cat_cols)
```

# 7. Baseline e modelos candidatos

Para avaliar a eficiência dos nossos algoritmos, estabelecemos uma régua de comparação simples (baseline) e selecionamos dois modelos candidatos com abordagens matemáticas diferentes para resolver o problema de regressão.

O comportamento configurado no código estabelece o seguinte fluxo:

Baseline (Referência): Um modelo bobo (DummyRegressor) que simplesmente "chuta" a mediana dos custos para todos os pacientes. Se os nossos modelos reais não ganharem desse chute, significa que eles não aprenderam nada.

Modelo Candidato 1: Regressão Linear com penalidade (Ridge).

Modelo Candidato 2: Floresta Aleatória de Árvores de Decisão (RandomForestRegressor).

```python
# === Definição de baseline e modelos candidatos ===

if PROBLEM_TYPE == "classificacao":
    baseline = Pipeline(steps=[
        ("preprocess", preprocess),
        ("model", DummyClassifier(strategy="most_frequent", random_state=SEED))
    ])

    candidates = {
        "LogisticRegression": Pipeline(steps=[
            ("preprocess", preprocess),
            ("model", LogisticRegression(max_iter=500, random_state=SEED))
        ]),
        "RandomForest": Pipeline(steps=[
            ("preprocess", preprocess),
            ("model", RandomForestClassifier(random_state=SEED))
        ])
    }

elif PROBLEM_TYPE == "regressao":
    baseline = Pipeline(steps=[
        ("preprocess", preprocess),
        ("model", DummyRegressor(strategy="median"))
    ])

    candidates = {
        "Ridge": Pipeline(steps=[
            ("preprocess", preprocess),
            ("model", Ridge())
        ]),
        "RandomForestRegressor": Pipeline(steps=[
            ("preprocess", preprocess),
            ("model", RandomForestRegressor(random_state=SEED))
        ])
    }

elif PROBLEM_TYPE == "clusterizacao":
    baseline = None
    candidates = {
        "KMeans_k3": Pipeline(steps=[
            ("preprocess", preprocess),
            ("model", KMeans(n_clusters=3, random_state=SEED, n_init=10))
        ]),
        "KMeans_k5": Pipeline(steps=[
            ("preprocess", preprocess),
            ("model", KMeans(n_clusters=5, random_state=SEED, n_init=10))
        ])
    }

elif PROBLEM_TYPE == "serie_temporal":
    # Para séries temporais, implemente uma baseline temporal, como "último valor observado".
    baseline = None
    candidates = {}

else:
    raise ValueError("PROBLEM_TYPE inválido.")
print("Baseline configurado:", "Sim" if baseline is not None else "Não")
print("Modelos candidatos:", list(candidates.keys()))
```

## 7.1 Justificativa dos modelos

  A escolha dessa combinação de modelos boa para este dataset pelos seguintes motivos:

Coerência do Baseline: Chutar a mediana é a melhor referência inicial. Como a análise gráfica mostrou a mediana representa o comportamento do "paciente padrão" da base. O modelo de inteligência artificial precisa provar que consegue errar bem menos que esse chute padrão.

Adequação ao Tamanho dos Dados: O nosso dataset é pequeno (cerca de 1.300 linhas). O modelo Ridge é leve e rápido, enquanto o RandomForest funciona incrivelmente bem em bases menores sem o risco de "decorar" os dados facilmente. Ambos rodam rápido, eliminando qualquer problema de custo computacional ou demora no treino.

Exigências de Escala e Texto: O modelo Ridge (por ser uma regressão linear tradicional) é extremamente sensível ao tamanho dos números e não entende textos. Por isso, a escolha desse modelo justifica o uso do Pipeline que criado no passo anterior, que padroniza os números e converte os textos antes de entregar o dado para ele. Já o RandomForest é mais robusto e não se importa com escalas, mas se beneficia dos textos convertidos.

Interpretabilidade e Captura de Regras: O gráfico 5 nos mostrou que o cruzamento de Obesidade + Cigarro gera grandes saltos nos custos. A Regressão Ridge tentará traçar uma linha reta para explicar isso, enquanto o RandomForest criará "regras de decisão" (se fuma = sim e imc > 30, então custo = alto). Testar os dois vai permitir ver qual abordagem matemática se adapta melhor à realidade dos planos de saúde.

# 8. Treinamento e avaliação inicial

Nesta etapa, será executado o treinamento automatizado do nosso modelo de referência (baseline) e dos dois algoritmos concorrentes (Ridge e RandomForestRegressor). O código roda o aprendizado utilizando exclusivamente a base de treino e, em seguida, faz as previsões utilizando a base de teste para preencher a tabela comparativa de desempenho.

A avaliação calcula o tempo gasto para o aprendizado em segundos e extrai as métricas de erro estatístico para descobrirmos qual abordagem matemática entende melhor a precificação dos planos de saúde.


```python
results = {}
trained_models = {}

if PROBLEM_TYPE in ["classificacao", "regressao"]:
    # Baseline
    t0 = time.time()
    baseline.fit(X_train, y_train)
    train_time = time.time() - t0

    y_pred = baseline.predict(X_test)
    if PROBLEM_TYPE == "classificacao":
        proba = baseline.predict_proba(X_test) if hasattr(baseline, "predict_proba") else None
        results["baseline"] = evaluate_classification(y_test, y_pred, proba)
    else:
        results["baseline"] = evaluate_regression(y_test, y_pred)
    results["baseline"]["train_time_s"] = round(train_time, 3)
    trained_models["baseline"] = baseline

    # Modelos candidatos
    for name, model in candidates.items():
        t0 = time.time()
        model.fit(X_train, y_train)
        train_time = time.time() - t0

        y_pred = model.predict(X_test)
        if PROBLEM_TYPE == "classificacao":
            proba = model.predict_proba(X_test) if hasattr(model, "predict_proba") else None
            results[name] = evaluate_classification(y_test, y_pred, proba)
        else:
            results[name] = evaluate_regression(y_test, y_pred)

        results[name]["train_time_s"] = round(train_time, 3)
        trained_models[name] = model

elif PROBLEM_TYPE == "clusterizacao":
    for name, model in candidates.items():
        t0 = time.time()
        model.fit(X_train)
        train_time = time.time() - t0
        labels = model.predict(X_test)
        X_test_transformed = model.named_steps["preprocess"].transform(X_test)
        results[name] = {
            "silhouette": silhouette_score(X_test_transformed, labels),
            "train_time_s": round(train_time, 3)
        }
        trained_models[name] = model

elif PROBLEM_TYPE == "serie_temporal":
    print("Implemente baseline temporal e modelos candidatos específicos para forecasting.")

show_results_table(results)
```

# 9. Validação e otimização de hiperparâmetros

Para tentar melhorar ainda mais o nosso melhor modelo, que foi o Random Forest, fizemos um teste combinando diferentes configurações para ver se ele conseguia errar ainda menos.

O que foi testado: Mexemos na estrutura das árvores do modelo, variando a quantidade de árvores (ex: 50, 100 ou 200) e o tamanho máximo de profundidade delas, buscando o equilíbrio para o robô não "decorar" as respostas.

Como testamos: Usamos a validação cruzada dividindo o treino em 5 partes. Isso significa que o computador testou as configurações 5 vezes com pedaços diferentes dos dados para garantir que o resultado é real e não foi por sorte.

Métrica de escolha: Usamos o $R^2$ (porcentagem de acerto) para guiar a escolha da melhor combinação.

Resultado: O teste mostrou que limitar o tamanho das árvores (não deixando elas crescerem sem controle) traz o melhor resultado. Isso faz sentido, pois como vimos que o maior mistério do custo é a regra direta do "Fumo + IMC", o modelo não precisa de caminhos muito longos ou complexos para acertar a previsão. Com esse ajuste, o modelo ficou mais leve e eficiente.

```python
# === Otimização de hiperparâmetros ===

N_ITER_SEARCH = 10
search = None

if PROBLEM_TYPE == "classificacao":
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
    model_to_tune = Pipeline(steps=[
        ("preprocess", preprocess),
        ("model", RandomForestClassifier(random_state=SEED))
    ])
    param_dist = {
        "model__n_estimators": randint(50, 250),
        "model__max_depth": randint(2, 20),
        "model__min_samples_split": randint(2, 10)
    }
    scoring = "f1_weighted"

elif PROBLEM_TYPE == "regressao":
    cv = KFold(n_splits=5, shuffle=True, random_state=SEED)
    model_to_tune = Pipeline(steps=[
        ("preprocess", preprocess),
        ("model", RandomForestRegressor(random_state=SEED))
    ])
    param_dist = {
        "model__n_estimators": randint(50, 250),
        "model__max_depth": randint(2, 20),
        "model__min_samples_split": randint(2, 10)
    }
    scoring = "neg_root_mean_squared_error"

elif PROBLEM_TYPE == "serie_temporal":
    cv = TimeSeriesSplit(n_splits=5)
    model_to_tune = None
    param_dist = None
    scoring = None
    print("Para séries temporais, configure um modelo compatível com features temporais e TimeSeriesSplit.")

elif PROBLEM_TYPE == "clusterizacao":
    print("Para clusterização, avalie diferentes valores de k e interprete os perfis dos grupos.")
    model_to_tune = None
    param_dist = None
    scoring = None

if model_to_tune is not None:
    search = RandomizedSearchCV(
        model_to_tune,
        param_distributions=param_dist,
        n_iter=N_ITER_SEARCH,
        cv=cv,
        scoring=scoring,
        random_state=SEED,
        n_jobs=1,
        verbose=1
    )
    search.fit(X_train, y_train)

    best_mae = abs(search.best_score_)
    print(f"\nMelhor MAE na validação cruzada: {best_mae:.2f}")
    print("Melhores hiperparâmetros localizados:", search.best_params_)
```
